```python
import os
import pandas as pd

def generate_metadata(image_dir, mask_dir, output_file="metadata.csv"):
    # Récupérer la liste des fichiers et les trier pour la correspondance
    images = sorted(os.listdir(image_dir))
    masks = sorted(os.listdir(mask_dir))

    # Filtrer pour ne garder que les fichiers (ignore les dossiers cachés)
    images = [f for f in images if os.path.isfile(os.path.join(image_dir, f))]
    masks = [f for f in masks if os.path.isfile(os.path.join(mask_dir, f))]

    # Vérification de sécurité
    if len(images) != len(masks):
        print(f" Attention : Nombre d'images ({len(images)}) différent du nombre de masques ({len(masks)})")

    # Création des listes de chemins (relatifs ou complets)
    
    data = {
        "Image": images,
        "Mask": masks
    }

    # Création du DataFrame et export
    df = pd.DataFrame(data)
    df.to_csv(output_file, index=False)
    
    print(f" Fichier '{output_file}' généré avec succès ({len(df)} lignes).")

# --- CONFIGURATION ---
PATH_IMAGES = "./Images"
PATH_MASKS = "./Masks"

generate_metadata(PATH_IMAGES, PATH_MASKS)
```
